```python
#@markdown # Sample plot with a Sine wave, Cosine wave and a Tangent wave

# prompt: plot with a sine,cos&tan wave

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

plt.plot(x, y_sin, label='sine')
plt.plot(x, y_cos, label='cosine')
plt.plot(x, y_tan, label='tangent')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine, Cosine, and Tangent Waves')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# @title Easing Functions Comparison
import matplotlib.pyplot as plt
import numpy as np

def ease_in_out_sine(t):
    return -(np.cos(np.pi * t) - 1) / 2

def ease_in_out_cubic(t):
    # Use NumPy's element-wise operations to apply the logic to each element in the array
    result = np.where(t < 0.5, 4 * t**3, (t - 1) * (2 * t - 2) * (2 * t - 2) + 1)
    return result

def ease_in_out_expo(t):
    # Use NumPy's where function for element-wise conditional logic
    return np.where(t == 0, 0, np.where(t == 1, 1, np.where(t < 0.5, np.power(2, 20 * t - 10) / 2, (2 - np.power(2, -20 * t + 10)) / 2)))


def ease_in_out_elastic(t):
    c4 = (2 * np.pi) / 3

    # Use NumPy's where function for element-wise conditional logic
    return np.where(t == 0, 0, np.where(t == 1, 1, np.where(t < 0.5, -(np.power(2, 20 * t - 10) * np.sin((20 * t - 11.125) * c4)) / 2, (np.power(2, -20 * t + 10) * np.sin((20 * t - 11.125) * c4)) / 2 + 1)))


t = np.linspace(0, 1, 100)

plt.plot(t, ease_in_out_sine(t), label='InOutSine')
plt.plot(t, ease_in_out_cubic(t), label='InOutCubic')
plt.plot(t, ease_in_out_expo(t), label='InOutExpo')
plt.plot(t, ease_in_out_elastic(t), label='InOutElastic')


plt.xlabel('t')
plt.ylabel('y')
plt.title('Easing Functions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# @title Random Polygons
# prompt: plot containing polygons at random x&y pos, which every point is on a random x&y pos, makr the color value go up to make an rainbow effect

import matplotlib.pyplot as plt
import numpy as np

# Number of polygons
num_polygons = 5

# Create a figure and axes
fig, ax = plt.subplots()

# Iterate through polygons
for i in range(num_polygons):
    # Number of points in each polygon (adjust as needed)
    num_points = np.random.randint(3, 40)  # Random number of points between 3 and 7

    # Generate random x and y coordinates for each point
    x = np.random.rand(num_points)
    y = np.random.rand(num_points)

    # Color based on the index for a rainbow effect
    color = plt.cm.hsv(i / num_polygons)  # Use HSV colormap for rainbow effect

    # Plot the polygon
    polygon = plt.Polygon(np.column_stack((x,y)), color=color)
    ax.add_patch(polygon)

# Set axis limits
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# Display the plot
plt.show()
```
